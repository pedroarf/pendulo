```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

teta=55 #ângulo inicial em graus
tempo=np.arange(0,10,0.001) #lista de tempo
g=9.8 #m/s**2
l=1.9 #comprimento da corda em metros

ci=[math.radians(teta),0] #condição incial

def DifPendulo(Y,t):
    teta=Y[0]
    omega=Y[1]
    dtetadt=omega
    domegadt=-g*(math.sin(teta))/l
    return (dtetadt, domegadt)

sol=odeint(DifPendulo,ci,tempo)

#passando para cordenadas cartesianas

Px=[]
Py=[]
for teta in sol[:,0]:
    x=l*math.sin(teta)
    Px.append(x)
    y=-l*math.cos(teta)
    Py.append(y)


    
plt.plot(tempo,sol[:,0])
plt.grid(True)
plt.title("Ângulo ao longo do tempo")
plt.xlabel("Tempo [segundos]")
plt.ylabel("Teta [radianos]")
plt.show()

a=sol[:,0][0]
i=0
while i <= len(sol[:,0]):
    if sol[:,0][i]>=a-0.00001 and i>200:
        print("Período [segundos]=")
        print(i/1000)
        print("Frequência [Hz]=")
        print((i/1000)**-1)
        break
    i=i+1
    
    
plt.plot(tempo,sol[:,1])
plt.grid(True)
plt.title("Velocidade angular ao longo do tempo")
plt.xlabel("Tempo [segundos]")
plt.ylabel("Omega [radianos/segundo]")
plt.show()

plt.plot(Px,Py)
plt.grid(True)
plt.title("Trajetória do pêndulo por x e y")
plt.xlabel("Distância em x [metros]")
plt.ylabel("Altura [metros]")
plt.show()
```
